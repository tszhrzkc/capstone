First I import pandas and use the 'read_html' method to get all the tables from the web page. And I select the first table which is what I need.

```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header = 0)
postal0 = df[0]
postal0
```

Now I drop all the rows whose 'Borough' equal to 'Not assigned' and reset the index.

```python
postal1 = postal0[postal0['Borough'] != 'Not assigned'].reset_index(drop = True)
postal1
```

Then I loop through the dataframe to get new content for rows whose 'Neighborhood' equal to 'Not assigned'.

```python
for index, (borough, neighborhood) in enumerate(zip(postal1.Borough, postal1.Neighborhood)):
    if neighborhood == 'Not assigned':
        postal1.Neighborhood[index] = borough
postal1
```

Now I get another csv file which contains all the geospatial data and merge this dataframe and postal1.

```python
df2 = pd.read_csv('http://cocl.us/Geospatial_data', header = 0)
postal2 = pd.merge(postal1, df2, on = ['Postal Code'])
postal2
```
